# 13. List Comprehensions
 * __List comprehension in Python is an easy and compact syntax for creating a list from a string or another list.__
 * __It is very concise way to create a new list by operating on each item in the existing list.__
 * __List comprehension is considerably faster than processing a lsit using the for loop.__
 * __Shorter and cleaner syntax for creating lists compared to loops:__
    * __[expression for items in iterable]__
 * __Can include conditions(if statements):__
    * __[expression for item in iterable if condition]__
 * __supports if-else expressions:__
    * __[a if condition else b for item in iterable]__
 * __Can replace simpole loops used for building lists.__
 * __Improves readability when used correctly.__
 * __Can handle nested loops:__
    * __[expression for x in list1 for y in list2]__
 * __Runs faster than traditional loops in most cases.__
 * __Should be used for simple operations to avoid making code hard to read.__
 * __Works with any iterable: lists, tuples, strings, ranges, etc.__

```python
# 1-10 

list1 = []
for i in range(1,11):
    # print(i,end="")
    list1.append(i)
print(list1)
```

```python
# [expression for variable in iterable]

[i for i in range(1,11)]
```

```python
# find out even number using list comprehensions.

[i for i in range(1,11) if i%2==0]
```

```python
[x for x in range(1,11) if x%2!=0]
```

```python
# [expression if condition else statement for i in iterable]

[x if x%2==0 else "Odd" for x in range(1,11)]
```

```python
["Even" if x%2==0 else x for x in range(1,11)]
```

```python
# display only c characters name.
languages = ['Python',"c++","JavaScript","C#","ForTran","BASIC","Go","Kotlin","C"]

[i for i in languages if "c" in i.lower()]
```

```python
for i in languages:
    if i[0].lower() == "c":
        print(i)
```

```python
[i for i in languages if i[0].lower() == "c"]
```

```python
# write a progam to display languages more than 5 characters.


[i for i in languages if len(i)>5]
```

```python
# Create a matrix (list of lists) of size 3Ã—3 containing numbers from 1 to 9 using nested list comprehension.

matrix = [[j for j in range(i, i+3)] for i in range(1, 10, 3)]
print(matrix)
```
