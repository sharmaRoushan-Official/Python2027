# 13. Classes and Objects in Python

 * __Python is an object-oriented programming structure language.__
 * __Unlike procedure-oriented progrmming, where the main emphasis is on functions, object-oriented programming stresses on objects.__
 * __An object is simply a collection of data(variables) and methods(functions) that act on those data.__
 * __Similarly, a class is bluprint for that object.__
 * __Like function definition begin with def keyword in Python, class definition begin with a class keyword.__
 * __The first string inside the class is called docstring and has a brief descrption of the class.__
 * __Although not mandatory, this is highly recommended.__
 * __Syntax:__
 * __class student:__
   * __school_name = "ABC School"__
   * __def _init_(self,name,age):__ constructor
     * __self.name = name__ # Run time Initialization
     * __self.age = age__
   * __obj = student("Harry",25)__
   * __obj.name__
   * __obj.age__

```python
# find the area of rectangle using class and object 

# mathematics 

# find the area of circle, circumference of circle. where radius is 10cm
# area of circle: pi*r**2 , where pi is 3.147
# circumference of cirlce: 2*pi*r
import math
class circle: # name of the class
    def __init__(self, radius): # constructor
        self.radius = radius   # instance   

    def areaCircle(self):
        area = math.pi*(self.radius**2)
        print(f"Area of Cirlce is: {area} cm²")
    def circumferenceCircle(self):
        circumference = 2*math.pi*self.radius
        print(f"Circumference of Circle: {circumference} cm")

    
circle1 = circle(7)
circle1.areaCircle()
circle1.circumferenceCircle()
```

```python
# find the area of triangle and their perimeter. side is 45,66,88


class rectangle:
    def __init__(self,length,width,height):
        self.length = length
        self.width = width
        self.height = height
    def detailsRectangle(self):
        area = (self.length)*(self.width)
        print(f"Area of Rectangle is: {area} cm2")

        perimeter = 2*((self.length)+(self.width))
        print(f"Perimeter of Rectangle is:{perimeter} cm")
        

rect1 = rectangle(45,66,88)
rect1.detailsRectangle()
```

```python
# find the area of parallelogram and thier perimeter.


# find the area of parallelogram and its perimeter.

class parallelogram:
    def __init__(self,length,width,height):
        self.length = length
        self.width = width
        self.height = height
    def detailsparallelogram(self):
        area = (self.length)*(self.width)
        print(f"Area of parallelogram is: {area} cm2")
        perimeter = 2*((self.length)+(self.width))
        print(f"Perimeter of parallelogram is:{perimeter} cm")

para1 = parallelogram(20,15,12)
para1.detailsparallelogram()
```

```python
# Hierarchical Inheritance
# one parent has multiple child

class a: # parent
    brandName1 = "Amazon"
    brandName2 = "Flipkart"
    brandName3 = "Meta"
    
class b(a): # child of a
    prod_1 = "Online Commerce Store"
    prod_2 = "Online Store"


class c(a): # child of a
    popularity1 = "100"
    popularity2 = "200"

class d(a):  # child of a
    x = 'apple'
    b = "pineApple"

obj1 = b()
print(obj1.brandName1,obj1.brandName2,obj1.brandName3, obj1.prod_1, obj1.prod_2)


obj2 = c()
print(obj2.brandName1, obj2.brandName2, obj2.brandName3, obj2.popularity1, obj2.popularity2)

obj3 = d()
print(obj3.brandName1, obj3.brandName2, obj3.brandName3, obj3.x, obj3.b)
```

```python
# Hybrid Inheritance

class a: # parent
    def roseBlack(self):
        return "Black Rose"

class b: # parent
    def roseLavender(self):
        return "Lavender Rose"

class c(a): # child of a 
    def purpleRose(self):
        return "Purple Rose"
class d(b,c): #child of b and c
    def childRose(self):
        return "HybridRose"


obj = d()
print(obj.roseBlack())
print(obj.roseLavender())
print(obj.purpleRose())
print(obj.childRose())
```

# Encapsulations 

 * __Encapsulation is an OOPs concept where we bundle the data (vairbales) and the methods (functions) that operate on that data into a single unit(class).__
 * __It is also direct access to some data- meaning you can hide sensitive information from the outside world and control access using getters and setters.__
 * __Real-life example: A bank account object conatains.__
 * __Data: Account Number, Balance, personal details__
 * __Method: Deposite, withdraw__
   * __The Account balance should not be directly modified by anyone, only through controlled methods.__
---
 * __Key Points About Encapsulations in Python__
 * __Public Members:Accessible Everywhere.__
 * __Private members: Declared with _ _ (Double underscore), not accessible directly outside the class.__
 * __Protected members: Declared with _ (single underscore), means to be treated as private but still accessible.__
 

```python
class BankAccount:
    def __init__(self, accountNumber,balance):
        self.accountNumber = accountNumber  # public attribute
        self.__balance = balance # private


    # Getter method  [Get only read the value]
    def get_balance(self):
        return self.__balance

    # setter method [update the value]
    def depositeAmount(self):
        amount = int(input("Enter your Amount ₹: "))
        if amount>0:
            self.__balance = self.__balance+amount
            print(f"Your Updated Amount is ₹: {self.__balance}")
        else:
            print(f"{amount} is Invalid Deposite Amount")

    def withdrawAmount(self):
        amount = int(input("Enter your Withdrawal Amount ₹: "))
        if 0<amount<=self.__balance: 
            self.__balance = self.__balance - amount
            print(f"Withdrawal amount is ₹{amount}\nYour Updated Balance is ₹{self.__balance}")
        else:
            print("Insufficient Funds or Invalid Amount")
            
smallBank = BankAccount("1234567899",100)
print(smallBank.accountNumber) # accessing outside the class
# print(smallBank.balance)  # not accessible outisde the class.
print(smallBank.get_balance())
```

```python
print(smallBank.depositeAmount())
```

```python
print(smallBank.withdrawAmount())
```

```python
# super() inheritance 

class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

obj = student('Linesh',35)
print(obj.name)
print(obj.age)
```

```python
class student2(student):  # name,age, email
    def __init__(self,name,age,email):
        self.email = email
        super().__init__(name,age)

obj3 = student2('Vivek',45,'vivek25@gmail.com')
print(obj3.name)
print(obj3.age)
print(obj3.email)
```

```python
class student3(student2):  # name,age,email,address
    def __init__(self,name,age,email,address):
        self.address = address
        super().__init__(name,age,email)

obj3 = student3('Vivek',35,'vivek@hotmail.com',"Texas")
print(obj3.name)
print(obj3.age)
print(obj3.email)
print(obj3.address)
```

```python
# 8.Demonstrate *operator overloading* by overloading the + operator for a class Point(x, y) to add coordinates using polymorphism.


class Point:
    def __init__(self,x,y):
        self.x = x  # 1
        self.y= y # 2

    def __add__(self,other):  
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(1,2)
print(p1.x,p1.y)

p2 = Point(3,4)
print(p2.x,p2.y)

p3 = p1+p2
print(p3.x,p3.y)
```
