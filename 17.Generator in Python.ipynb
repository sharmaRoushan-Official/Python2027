# 17. Generator in Python 

 * __Python generators are function that retuns the traversale object and a simple way of creating iterators.__
 * __It traverses all the items at once.__
 * __The generator can also be expressed in a syntax similar to the list.__
 * __There is a lot of complexity in creating iteration in Python. It is required to implement iter() and next() methods to keep track of internal states.__
 * __It is lengthy process to create iterators: it raise stopIteration exceptions.__
 * __It pauses the function execution by saving all states and yield to the caller.__
 * __Later, it resumes execution when a successive function is called.__

```python
# 1 - 10
for i in range(1,11):
    print(i)
```

```python
i = 1
j = 11
while i<j:
    print(i)
    i+=1
```

```python
def func(n=20):
    for i in range(1,n+1):
        print(i)
func(25)
```

```python
def checkval(x=50):
    if x%2==0:
        yield x
checkval()
```

```python
def message():
    oneMessage = "Hello world"
    return oneMessage

def message2():
    twoMessage = "Congratulations! Linesh"
    return twoMessage

def message3():
    threeMessage = "Hey Daya"
    return threeMessage

print(message())
print(message2())
print(message3())
```

```python
def message():
    oneMessage = "Hello world"
    yield oneMessage


    twoMessage = "Congratulations! Linesh"
    yield twoMessage


    threeMessage = "Hey Daya"
    yield threeMessage

message()
```

```python
# Create a generato that yields number from 1 to n

def count_up(n):
    for i in range(1,n+1):
        yield i

print(list(count_up(10)))
```

```python
def count_up(n):
    for i in range(1,n+1):
        if i%2==0:
            yield i

print(list(count_up(10)))
```

```python
def count_up(n):
    for i in range(1,n+1):
        if i%2!=0:
            yield i

print(list(count_up(10)))
```

```python
# infinite generator

# write an inifinite generator for natural numbers

def inifinite_numbers():
    num = 1
    while True:
        yield num
        num+=1

x = inifinite_numbers()
print(next(x))
```
