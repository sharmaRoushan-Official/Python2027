```python
from datetime import datetime, timedelta , UTC
def getWeather():
    api_key = "1936a33cac4337dc2ec5eed9bbaa2f1c"
    cityName1 = input("Enter City Name: ")
    countryCode = input("Enter Country Name: ")
    cityName = cityName1+","+countryCode
    url = f"https://api.openweathermap.org/data/2.5/weather?q={cityName}&appid={api_key}"
    print(url)

    res = requests.get(url)
    print(res) #status of the code

    if res.status_code == 200:
        data = res.json() #Python dictionary: json_object
        print(data)

        temp = data['main']['temp']
        temp1 = data['main']['feels_like']
        tempKtoC = temp - 273.15
        feelsLike = temp1- 273.15
        humidity = data['main']['humidity']
        pressure = data['main']['pressure']
        weather_desc = data['weather'][0]['description']
        wind_speed = data['wind']['speed']
        visibility = data.get('visibility', 0) / 1000  # meters → km

        
        dateTime = data['timezone']
        # print(dateTime)
        utc_now = datetime.now(UTC)
        local_time = utc_now + timedelta(seconds = data['timezone'])
        # print(local_time)

        print("\nWeather Details")
        print("----------------")
        
        print(f"Current Time in {data['name']}, {local_time.strftime('%Y-%m-%d %H:%M:%S %p')}")
        print("Todays Temperature is: ",tempKtoC,"°C")
        print("Temperature Feels Like: ",feelsLike,"°C")
        print("Condition:", weather_desc.capitalize())
        print("Humidity:", humidity, "%")
        print("Pressure:", pressure, "hPa")
        print("Wind Speed:", wind_speed, "m/s")
        print("Visibility:", visibility, "km")
     
    else:
        print("Failed to fetch weather data")    

getWeather()
```
