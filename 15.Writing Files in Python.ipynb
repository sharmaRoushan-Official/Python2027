# 15. Writing Files in Python

 * __To Write a text file in Python, you follow these stesp:__
 * __1.First, Open the text for writing (or appending) using the open function.__
 * __2.Second, Write to the text file using write(), or writelines() method.__
 * __3.Third, close the file using close() method.__

### Some characters and thier functions
 * __"r": Open file for reading only. Starts reading from the beginning of the file. This is default mode.__
 * __"rb": Open file for reading only in binary format. Starts reading from the beginning of the text.__
 * __"r+": Open file for reading and writing. File pointer places at beginning of the file.__
 * __"w": Open file for writing only. File pointer places at beginning of the file. Overwirtes an exising file and creates a new one if it does exits.__
 * __"wb":Same as W but opens in binary mode.__
 * __"wb+": Same as wb but also allows reading text from the file.__
 * __"a": Open file for appending. Starts writing at the end of the file. creates a new file if does not exits.__
 * __"ab": Same as a but in binary format. Create a new file if files does not exits.__
 * __"ab+" Same as ab but also open for reading.__

```python
# using try and except write a files 

fileName = input("Enter your file name here: ") # linesh

completeFileName = "C:\\Users\\User\\Desktop\\"+fileName+".txt"


with open(completeFileName,'w') as f:
    try:
        lines = int(input("Enter Number of lines: "))
        for i in range(1,lines+1):
            content = input("Start Writing Here: ")
            f.write(str(i))
            f.write('.')
            f.write(content)
            f.write('\n')
    except Exception as e:
        print(e)
```

```python
# writes a file to count total number of character with space and without space and display in the text file.

fileName = input("Enter your file name here: ")
completeFileName = "C:\\Users\\User\\Desktop\\"+fileName+".txt"

with open(completeFileName, 'w') as f:
        lines = int(input("Enter Number of lines: "))
        for i in range(1, lines + 1):
            content = input("Start Writing Here: ")
            f.write(str(i))
            f.write('.')
            f.write(content)
            f.write('\n')

            
with open(completeFileName, 'r') as f:
    for i in f:
        countChar = 0
        space = 0
        for j in i:
            countChar += 1
            if j == " ":
                space += 1

        charWithout_space = countChar - space
    with open(completeFileName, 'w') as f:
        f.write("\nTotal Characters with space: " + str(countChar))
        f.write("\nTotal Characters without space: " + str(charWithout_space))
```
