# 14. Reading Files in Python 

 * __To read a text files in Python, you follow these steps.__
 * __First, Open a text file for reding by using the open() function.__
 * __Second, read text from the text file using read(), readline(), readlines() method of the file object.__
 * __Third, close the file using the file close() method, which frees up resources and ensures consistency across different Python version.__
 * __Method ------------------ |Descrpitons|__
 * __Writable(): Returns whether the file can be written or not.__
 * __Readable(): Returns whether the file stream can be read or not.__
 * __Readlines():Returns one line form the file.__
 * __Write():Write the spacified string to the file.__
 * __Writelines(): Writes a list of strings to the file.__
