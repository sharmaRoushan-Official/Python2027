# 13. Classes and Objects in Python

 * __Python is an object-oriented programming structure language.__
 * __Unlike procedure-oriented progrmming, where the main emphasis is on functions, object-oriented programming stresses on objects.__
 * __An object is simply a collection of data(variables) and methods(functions) that act on those data.__
 * __Similarly, a class is bluprint for that object.__
 * __Like function definition begin with def keyword in Python, class definition begin with a class keyword.__
 * __The first string inside the class is called docstring and has a brief descrption of the class.__
 * __Although not mandatory, this is highly recommended.__
 * __Syntax:__
 * __class student:__
   * __school_name = "ABC School"__
   * __def _init_(self,name,age):__ constructor
     * __self.name = name__ # Run time Initialization
     * __self.age = age__
   * __obj = student("Harry",25)__
   * __obj.name__
   * __obj.age__

```python
# find the area of rectangle using class and object 

# mathematics 

# find the area of circle, circumference of circle. where radius is 10cm
# area of circle: pi*r**2 , where pi is 3.147
# circumference of cirlce: 2*pi*r
import math
class circle: # name of the class
    def __init__(self, radius): # constructor
        self.radius = radius   # instance   

    def areaCircle(self):
        area = math.pi*(self.radius**2)
        print(f"Area of Cirlce is: {area} cm²")
    def circumferenceCircle(self):
        circumference = 2*math.pi*self.radius
        print(f"Circumference of Circle: {circumference} cm")

    
circle1 = circle(7)
circle1.areaCircle()
circle1.circumferenceCircle()
```

```python
# find the area of triangle and their perimeter. side is 45,66,88


class rectangle:
    def __init__(self,length,width,height):
        self.length = length
        self.width = width
        self.height = height
    def detailsRectangle(self):
        area = (self.length)*(self.width)
        print(f"Area of Rectangle is: {area} cm2")

        perimeter = 2*((self.length)+(self.width))
        print(f"Perimeter of Rectangle is:{perimeter} cm")
        

rect1 = rectangle(45,66,88)
rect1.detailsRectangle()
```

```python
# find the area of parallelogram and thier perimeter.


# find the area of parallelogram and its perimeter.

class parallelogram:
    def __init__(self,length,width,height):
        self.length = length
        self.width = width
        self.height = height
    def detailsparallelogram(self):
        area = (self.length)*(self.width)
        print(f"Area of parallelogram is: {area} cm2")
        perimeter = 2*((self.length)+(self.width))
        print(f"Perimeter of parallelogram is:{perimeter} cm")

para1 = parallelogram(20,15,12)
para1.detailsparallelogram()
```

```python
# Multiple Inheritance

class a: # parent
    brandName1 = "Amazon"
    brandName2 = "Flipkart"
    brandName3 = "Meta"
    
class b: #parent
    prod_1 = "Online Commerce Store"
    prod_2 = "Online Store"


class c: #parent
    popularity1 = "100"
    popularity2 = "200"

class d(a,b,c):  # one child has multiple parent
    x = 'apple'
    b = "pineApple"

obj1 = d()
print(obj1.brandName1,obj1.brandName2, obj1.brandName3, obj1.popularity1, obj1.popularity2, obj1.prod_1,obj1.prod_2, obj1.x, obj1.b)
```

```python
# Hierarchical Inheritance
# one parent has multiple child

class a: # parent
    brandName1 = "Amazon"
    brandName2 = "Flipkart"
    brandName3 = "Meta"
    
class b(a): # child of a
    prod_1 = "Online Commerce Store"
    prod_2 = "Online Store"


class c(a): # child of a
    popularity1 = "100"
    popularity2 = "200"

class d(a):  # child of a
    x = 'apple'
    b = "pineApple"

obj1 = b()
print(obj1.brandName1,obj1.brandName2,obj1.brandName3, obj1.prod_1, obj1.prod_2)


obj2 = c()
print(obj2.brandName1, obj2.brandName2, obj2.brandName3, obj2.popularity1, obj2.popularity2)

obj3 = d()
print(obj3.brandName1, obj3.brandName2, obj3.brandName3, obj3.x, obj3.b)
```

```python
# Hybrid Inheritance

class a: # parent
    def roseBlack(self):
        return "Black Rose"

class b: # parent
    def roseLavender(self):
        return "Lavender Rose"

class c(a): # child of a 
    def purpleRose(self):
        return "Purple Rose"
class d(b,c): #child of b and c
    def childRose(self):
        return "HybridRose"


obj = d()
print(obj.roseBlack())
print(obj.roseLavender())
print(obj.purpleRose())
print(obj.childRose())
```

```python
# Polymorphism 
# Overriding

class Animal:
    def speak(sefl):
        print("Animal Speak")

class dog(Animal):
    def speak(self):
        return "Bark"


class cat(Animal):
    def speak(self):
        return "Meaw, Meaw, Meaw"


class goat(Animal):
    def speak(self):
        return "Bleat"

dog1 = dog()
print(dog1.speak())

cat1 = cat()
print(cat1.speak())

goat1 = goat()
print(goat1.speak())
```
