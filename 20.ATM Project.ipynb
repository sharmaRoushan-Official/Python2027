```python
class atm:
    def __init__(self,name,accountNumber,pin,balance=0):
        self.name = name
        self.accountNumber=accountNumber
        self.pin = pin
        self.balance = balance
        


    def details(self):
        # print('Hello world')
        '''Display Details of Customer:'''
        print(f"Account Holder Name: {self.name}")
        print(f"Your Account Number: {self.accountNumber}")
        print(f"Your Current Balance: {self.balance}")

    def displayBalance(self):
        print(f"Your Current Balance is:{self.balance}")


    def depositeCash(self):
        pin2 = input("Enter your current PIN: ")
        if pin2 == self.pin:
            depo = int(input("Enter your amount: "))
            if depo>0:
                self.balance = self.balance+ depo
                print(f"Balance Deposited Successfully:₹ {depo}\nYour Updated Balance is:₹ {self.balance}")
            else:
                print("Invalid Amount. Enter Valid Amount")
        else:
            print("Invalid PIN. Enter Correct PIN and Try Again!!")


    def trans(self):
        print("""
        Select Options:
        \n1.Account Details
        \n2.Account Balance
        \n3.Deposite Cash
        \n4.Withdraw Balance
        \n5.Change Name
        \n6.ShowAccount Number
        \n7.Change PIN
        \n8.Exit
        \n----------------------------
        """)
        while True:
            try:
                opt = int(input("Choose from any 1-8 Options: "))
            except Exception as e:
                print("Error:",e)
                continue
            else:
                if opt == 1:
                    atm.details(self)
                elif opt == 2:
                    atm.displayBalance(self)
                elif opt == 3:
                    atm.depositeCash(self)
                elif opt == 8:
                    break
                    

cus1=atm("Linesh","123456789","1234",100)
cus1.trans()
```
