# 10. Exceptional Handling in Python

 * __An exception is an event, which occurs duing the execution of a program ther disrupts the normal flow of the program's instructions.__
 * __In general, when a python script encounters that it cannot cope with it, it raises an exceptions.__
 * __An exceptions is a python object that represents an error.__
 * __When a python script raises an exceptions, it must either handle the exception immediately otherwise it terminates an quits.__
 * __If you have some suspicious code that may raise an exceptions you can defend your program by placing the suspisious code in try:block.__
 * __After that try:block, include an except statement followed by a block of code which handles the problem as elegantly as possible.__
 * __Some common Exceptions.__
    * __ZeroDivisionError__
    * __ValueError__
    * __IOError[Input Output Error]__
    * __EOFError[End of File Error]__
    * __IndentationError__

```python
# wap to take some string from user and print value only with starts from "A"
# students = ["Ankita","anup", "Omkar","Manisha","Utkarsh","Neeraj","Atharv","Aman"]

try:
    listName = []
    studentStartsA = []
    def startA():
        while True:
            try:
                userName = input("Enter students Name: ")
            except Exception as e:
                print(e)
            if userName == "0":
                break
            listName.append(userName)

            if userName[0] == "A" or userName[0] == 'a':
                studentStartsA.append(userName)
            else:
                pass
    startA()
except Exception as e:
    print(e)
else:
    print("Generated List is: ",listName)
    print("Student Name Starts from A is:",studentStartsA)
finally:
    print("Program Executed Successfully.")
```

```python
# decimal - 0-9=10


# Write code that divides two numbers entered by the user.
try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))

    divNum = num1 / num2

except ValueError:
    print("This code gives an Error ValueError")

except ZeroDivisionError:
    print("This code gives an Error ZeroDivisionError")

else:
    print(f"The division of {num1} and {num2} is: {divNum}")
```
